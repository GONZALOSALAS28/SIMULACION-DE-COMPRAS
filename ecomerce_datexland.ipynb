```python
cities =['Buenos Aires', 'Córdoba', 'Rosario', 'Mendoza', 'Tucumán']
payment_online = ['Credit_card','PSE']
payment_store = ['Cash','banco provincia','mercado pago','Credit_card']
source = ['Facebook','Instagram','Organic','Twitter','Influencer_1','Influencer_2','Influencer_3','Influencer_4']
status_purchase = ['COMPLETED','FAILED_CHECKOUT','FAILED_API_RESPONSE','INSUFICCIENT_FUNDS','COMPLETED','COMPLETED','COMPLETED','COMPLETED','COMPLETED','COMPLETED','FAILED_API_RESPONSE','INSUFICCIENT_FUNDS','USER_ERROR','FRAUD','COMPLETED','COMPLETED','COMPLETED']
```

```python
buenos_aires_coords = [(-34.603641738779984, -58.411054077577475), (-34.603685839199976, -58.41054982260622), (-34.602767420204785, -58.41104334903878)]
cordoba_coords = [(-31.375011335790152, -64.20097072065562), (-31.377235618226326, -64.2389398521822)]
rosario_coords = [(-32.94665254651427, -60.63422671686838)]
mendoza_coords = [(-32.87055685331168, -68.84782888343456), (-32.876716840604466, -68.85302544621048)]
tucuman_coords = [(-26.83029669890043, -65.20458909172736), (-26.82723872462177, -65.20736761205426)]

```

```python
random.choice(payment_store)
```

```python
def get_pay_method(source,status_purchase,payment_online,payment_store):
    
    if source == 'Organic' :
        
        payment = random.choice(payment_store)
        status = 'COMPLETED'
        order_type = 'STORE'
    
    elif source != 'Organic' :
        
        payment = random.choice(payment_online)
        status = random.choice(status_purchase)
        order_type = 'ONLINE'
    
    return payment,status,order_type   

def get_coords(city):
    if city == 'Buenos Aires':
        return buenos_aires_coords[random.randint(0, len(buenos_aires_coords) - 1)]
    elif city == 'Córdoba':
        return cordoba_coords[random.randint(0, len(cordoba_coords) - 1)]
    elif city == 'Rosario':
        return rosario_coords[random.randint(0, len(rosario_coords) - 1)]
    elif city == 'Mendoza':
        return mendoza_coords[random.randint(0, len(mendoza_coords) - 1)]
    elif city == 'Tucumán':
        return tucuman_coords[random.randint(0, len(tucuman_coords) - 1)]
    else:
        return None



```

```python
x = 0

data_purchase = []

while(x < 10):
    
    date = pd.to_datetime('today').strftime("%Y-%m-%d %H:%M:%S")
    
    PRODUCTO = df['PRODUCTO '][random.randint(0,14)]
    PRECIO =  df[df['PRODUCTO ']==PRODUCTO]['PRECIO'].values[0]
    COMISION =  df[df['PRODUCTO ']==PRODUCTO]['COMISION'].values[0]
    MARCAS = df[df['PRODUCTO ']==PRODUCTO]['MARCAS'].values[0]
    CATEGORIA = df[df['PRODUCTO ']==PRODUCTO]['CATEGORIA'].values[0]
    source_temp = random.choice(source)
    pay = get_pay_method(source_temp,status_purchase,payment_online,payment_store)
    city = random.choice(cities)
    
    purchase = {'purchase_ID': str(hashlib.sha256(f"{x} {PRODUCTO} {PRECIO} {COMISION} {date} {source_temp} {pay[1]}".encode('utf-8')).hexdigest())[:10], 
            'PRODUCTO' : PRODUCTO,
            'PRECIO':str(PRECIO),
            'COMISION':str(COMISION),
            'Revenue' : str(round(PRECIO*COMISION,2)),
            'Payment_Mehtod':pay[0],
            'Status' : pay[1],
            'Order_Type' : pay[2],
            'City':city,
            'Latitud' : str(get_coords(city) [0]) ,
            'Longitud' :  str(get_coords(city)[1]),
            'Source':source_temp,
            'marcas' : MARCAS,
            'CATEGORIA': CATEGORIA,
            'Created_at': date}
    
    data_purchase.append(pd.DataFrame(purchase,index=[x]))
    print(purchase)
    


    x += 1
    time.sleep(random.choice([1,2]))
```

```python
df_final =  pd.concat(data_purchase,ignore_index=True)
df_final
```
